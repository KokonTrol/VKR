```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
from datetime import datetime
from scipy import stats
import matplotlib.pyplot as plt
```

```python
from matplotlib.pyplot import figure

figure(figsize=(7, 5), dpi=80)

import tkinter.filedialog as tkFileDialog
def open_file():
    return tkFileDialog.askopenfilename( title='Выберите файл:')
```

```python
data = pd.read_csv(open_file(), delimiter=";", encoding='cp1251', decimal=",") #delimiter=";", 
```

```python
corr = data[data.columns.difference(["Команда", "ФИО", "Баллы"])].corr()
corr.style.background_gradient(cmap='coolwarm')
# print(corr)
```

```python
corr = data[data.columns.difference(['ФИО', "Команда"])].corr()
corr.style.background_gradient(cmap='coolwarm')
# print(corr)
```
